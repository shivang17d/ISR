# Importing the necessary libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# URL 

```python
baseurl = "https://www.flipkart.com/laptops/~buyback-guarantee-on-laptops-/pr?sid=6bo%2Cb5g&uniqBStoreParam1=val1&wid=11.productCard.PMU_V2"
```

# Header

```python
#Creating a user agent to make every request (Default user agent may get blocked due to various security settings of the website)

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

# Request

```python
req = requests.get(baseurl).text
soup=BeautifulSoup(req,'html.parser')
print(soup)
```

# Concating reference URL with BaseURL

```python
url_list = soup.find_all('a')
for url in url_list:
  link = url.get('href')
  print(baseurl + link)
```

# Lists to store name,price,rating of product

```python
products=[] 
prices=[]
ratings=[] 
```

```python
product_list = soup.find_all("div",{"class":"_4rR01T"})
for product in product_list:
  products.append(str(product.text))
```

```python
price_list = soup.find_all("div",{"class":"_30jeq3 _1_WHN1"})
for price in price_list:
  p = price.text
  prices.append(p)
 
```

```python
rating_list = soup.find_all("div",{"class":"_3LWZlK"})
for rating in rating_list:
  ratings.append(str(rating.text))
```

# Displaying Results

```python
print("\t\t\tProduct Name" + "\t\t\t\t\t\t\t     "+ "Product price"+"\t "+ "Product rating")
for i in range(0,len(products)):
  print(products[i],end="\t\t")
  print(prices[i],end="\t\t\t")
  print(ratings[i])
  print("\n")
  
```
