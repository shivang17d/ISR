# Importing an Image :-

```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
image = mpimg.imread("puppy_image.jpeg")
plt.imshow(image)
plt.show()
```

# Red and Yellow image :-

```python
red,yellow=image.copy(),image.copy()
red[:,:,(1,2)]=0
yellow[:,:,2]=0
```

```python
plt.title('Red Intensity')
plt.imshow(red)
plt.show()
```

```python
plt.title('Yellow Intensity')
plt.imshow(yellow)
plt.show()
```

# Converting Images to a 2-D matrix (RGB TO Gray) :-

```python
from skimage.color import rgb2gray
gray_image=rgb2gray(image)
plt.title('Color Image')
plt.imshow(image)
plt.show()
plt.title('GrayScale')
plt.imshow(gray_image)
plt.show()
print ("Colored image shape: ",image.shape)
print ("Gray image shape: ",gray_image.shape)
```

# Binarization of gray scale image:- 

```python
from skimage.filters import threshold_otsu

thresh = threshold_otsu(gray_image)

binary = gray_image > thresh
plt.title('Gray Image')
plt.imshow(gray_image)
plt.show()
plt.title('Binary Image')
plt.imshow(binary)
plt.show()
```

# Blurring an Image:-

```python
from skimage.filters import gaussian
blurred_image=gaussian(gray_image,sigma=20)
plt.title('Gray Image')
plt.imshow(gray_image)
plt.show()
plt.title('20 sigma Blurr')
plt.imshow(blurred_image)
plt.show()
```
